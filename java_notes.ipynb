```java
// Methods of Collection Interface


add(Object o) : Adds an object to the collection.
addAll(Collection c) : Adds all the elements of the specified collection to the collection.
remove(Object o) : Removes an object from the collection.
removeAll(Collection c) : Removes all the elements of the specified collection from the collection.
clear() : Removes all the elements from the collection.
contains(Object o) : Checks whether the specified object is present in the collection or not.
containsAll(Collection c) : Checks whether all the elements of the specified collection are present in the collection or not.
isEmpty() : Checks whether the collection is empty or not.




// Methods of List Interface
// add(T type)
// add(int index, T type)
// remove(int index)
// remove(Object o)  // (Integer.valueOf(x));
// get(int index)
// set(int index, E element)
// int indexOf(Object o)
// int lastIndexOf(Object o)
add(int index, Object element) : Inserts the specified element at the specified position in the list.
addAll(int index, Collection c) : Inserts all the elements of the specified collection at the specified position in the list.
get(int index) : Returns the element at the specified position in the list.
set(int index, Object element) : Replaces the element at the specified position in the list with the specified element.
indexOf(Object o) : Returns the index of the first occurrence of the specified element in the list.
lastIndexOf(Object o) : Returns the index of the last occurrence of the specified element in the list.
listIterator() : Returns a list iterator over the elements in the list.
listIterator(int index) : Returns a list iterator over the elements in the list, starting at the specified position.
remove(int index) : Removes the element at the specified position in the list.
subList(int fromIndex, int toIndex) : Returns a view of the portion of the list between the specified fromIndex, inclusive, and toIndex, exclusive.



// Methods of Set Interface
add(Object o) : Adds the specified element to the set.
addAll(Collection c) : Adds all the elements of the specified collection to the set.
remove(Object o) : Removes the specified element from the set.
removeAll(Collection c) : Removes all the elements of the specified collection from the set.
clear() : Removes all the elements from the set.
contains(Object o) : Checks whether the specified element is present in the set or not.
containsAll(Collection c) : Checks whether all the elements of the specified collection are present in the set or not.
isEmpty() : Checks whether the set is empty or not.

// Methods of Stack Interface
push(Object o) : Pushes an item onto the top of the stack.
pop() : Removes the object at the top of the stack and returns that object as the value of this function.
peek() : Looks at the object at the top of the stack without removing it from the stack.

// Methods of Queue Interface
add(Object o) : Inserts the specified element into the queue.
offer(Object o) : Inserts the specified element into the queue.
remove() : Retrieves and removes the head of the queue.
poll() : Retrieves and removes the head of the queue.
element() : Retrieves, but does not remove, the head of the queue.
peek() : Retrieves, but does not remove, the head of the queue.




// Methods of Map Interface
put(Object key, Object value) : Associates the specified value with the specified key in the map.
putAll(Map m) : Copies all the mappings from the specified map to the map.
remove(Object key) : Removes the mapping for the specified key from the map.
clear() : Removes all the mappings from the map.
containsKey(Object key) : Checks whether the map contains a mapping for the specified key or not.
containsValue(Object value) : Checks whether the map contains a mapping for the specified value or not.
isEmpty() : Checks whether the map is empty or not.



// Methods of Iterator Interface
hasNext() : Returns true if the iteration has more elements.
next() : Returns the next element in the iteration.
remove() : Removes from the underlying collection the last element returned by the iterator.



// Methods of ListIterator Interface
hasPrevious() : Returns true if the list iterator has more elements when traversing the list in the reverse direction.
previous() : Returns the previous element in the list and moves the cursor position backwards.
previousIndex() : Returns the index of the element that would be returned by a subsequent call to previous().
set(Object o) : Replaces the last element returned by next() or previous() with the specified element.
add(Object o) : Inserts the specified element into the list.



// Methods of Enumeration Interface
hasMoreElements() : Returns true if the enumeration has more elements.
nextElement() : Returns the next element of the enumeration.


// Method of Collection itself
Collections.reverse(Object o) : Reverses the order of the elements in the specified list.
Collections.shuffle(Object o) : Randomly permutes the specified list using a default source of randomness.
Collections.sort(Object o) : Sorts the specified list into ascending order, according to the natural ordering of its elements.
Collections.swap(Object o, int i, int j) : Swaps the elements at the specified positions in the specified list.
Collections.fill(Object o, Object obj) : Replaces all of the elements of the specified list with the specified element.
Collections.copy(Object dest, Object src) : Copies all of the elements from one list into another.
Collections.max(Object o) : Returns the maximum element of the specified collection, according to the natural ordering of its elements.
Collections.min(Object o) : Returns the minimum element of the specified collection, according to the natural ordering of its elements.
Collections.replaceAll(Object o, Object oldVal, Object newVal) : Replaces all occurrences of one specified value in a list with another.
Collections.frequency(Object o, Object obj) : Returns the number of elements in the specified collection equal to the specified object.


```

```java
// Iterating over Collections

// Iterating over List
List<String> names = new ArrayList<>(Arrays.asList("Clementine", "Duran", "Mike"));

names.forEach(System.out::println);    // Version ≥ Java SE 8


for (String name : names) {           // Version ≥ Java SE 5
    System.out.println(name);
}


for (Iterator<String> iterator = names.iterator(); iterator.hasNext(); ) { // Version ≥ Java SE 5
    System.out.println(iterator.next());
}

//Creates ListIterator which supports both forward as well as backward traversel
ListIterator<String> listIterator = names.listIterator();
//Iterates list in forward direction
while(listIterator.hasNext()){
    System.out.println(listIterator.next());
}


//Iterates list in backward direction once reaches the last element from above iterator in forward direction
while(listIterator.hasPrevious()){
    System.out.println(listIterator.previous());
}



// Iterating over Set
Set<String> names = new HashSet<>(Arrays.asList("Clementine", "Duran", "Mike"));

names.forEach(System.out::println);   //Version ≥ Java SE 8

for (Iterator<String> iterator = names.iterator(); iterator.hasNext(); ) {
    System.out.println(iterator.next());                                      //Version ≥ Java SE 5
}

for (String name : names) {
    System.out.println(name);
}


// Iterating over Map
Map<Integer, String> names = new HashMap<>();

names.forEach((key, value) -> System.out.println("Key: " + key + " Value: " + value));  //Version ≥ Java SE 8

for (Map.Entry<Integer, String> entry : names.entrySet()) {                             //Version ≥ Java SE 5
    System.out.println(entry.getKey());
    System.out.println(entry.getValue());
}

// Iterating over only keys
for (Integer key : names.keySet()) {
    System.out.println(key);
}

// Iterating over only values
for (String value : names.values()) {
    System.out.println(value);
}


Iterator entries = names.entrySet().iterator();
while (entries.hasNext()) {
    Map.Entry entry = (Map.Entry) entries.next();
    System.out.println(entry.getKey());
    System.out.println(entry.getValue());
}
```
